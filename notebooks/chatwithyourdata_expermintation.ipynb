## Load Environment Variables

```python
# load .env file, and get AI_SEARCH_ENDPOINT value
from dotenv import load_dotenv
import os

load_dotenv()
AI_SEARCH_ENDPOINT = os.environ["AI_SEARCH_ENDPOINT"]
```

## Experimentation with the [Chat with Your Data](https://github.com/Azure-Samples/chat-with-your-data-solution-accelerator) `api/conversation`

```python
# send a REST POST request to ENDPOINT
# - header has content type application/json
# - json body is using the contents of ai_payload.json
import requests
import json

with open("ai_payload.json", "r") as f:
    payload = json.load(f)
    
headers = {
    "Content-Type": "application/json"
    }
response = requests.post(AI_SEARCH_ENDPOINT, headers=headers, json=payload)

# print the response status code and text
print(response.status_code)
print(response.text)
```

```python
response_json = response.json()
last_message_content = response_json['choices'][-1]['messages'][-1]['content']

print(last_message_content)
```
